```python
import pandas as pd
```

```python
dados = pd.read_json('/content/dadosHospedagem.json') # Lê um arquivo .json a partir de um caminho especificado, convertendo seus dados para um Data Frame #
dados.head()
```

```python
dados = pd.json_normalize(dados['info_moveis']) # Como os dados estão aninhados, desaninha os dados, agrupando-os em listas #
```

```python
dados
```

##Retirando as Listas


```python
colunas = list(dados.columns) # Cria uma lista com os nomes das colunas #
colunas
```

```python
dados = dados.explode(colunas[3:]) # Desagrupa os dados nos índices (Nomes de colunas) especificados #
dados
```

```python
dados.reset_index(inplace=True,drop=True) # Reseta a coluna que enumera as linhas,(inplace=True) altera aplica a alteração diretamente à variável e (drop=True) faz com que a antiga coluna seja apagada #
dados.head()
```

```python
dados.info()
```

# Convertendo Dados Numéricos


```python
import numpy as np
```

```python
dados['max_hospedes']=dados['max_hospedes'].astype(np.int64)
```

```python
dados.info()
```

```python
colNumericas = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
```

```python
dados[colNumericas] = dados[colNumericas].astype(int)
```

```python
dados.info()
```

```python
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(float)
```

```python
dados.info()
```

```python
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$','').replace(',','').strip())
```

```python
dados['preco'] = dados['preco'].astype(float)
```

```python
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].applymap(lambda x: x.replace('$','').replace(',','').strip())
```

```python
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].astype(float)
```

```python
dados.info()
```

#Dados Textuais

```python
dados['descricao_local'] = dados['descricao_local'].str.lower() # Passa todo o texto como string e deixa maiúsculo #
```

```python
dados.head()
```

```python
dados['descricao_local'][1743]
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']',' ',regex=True) # Parâmetro Regex dentro de .replace() ignora os caracteres especificados após o '^' e substitui #
                                                                                                   # pelo parâmetro especificado após a primeira vírgula #
```

```python
dados['descricao_local'].str.replace('(?<!\w)-(?!\w)','',regex=True) # Lookbehind negativo (?<!) e Lookahead Negativo (?!) verificam a existência de caracteres antes e depois do hífen (-), #
                                                                     # se não houver, será substituído pelo espaço em branco depois da vírgula #
```

```python
dados['descricao_local'] = dados['descricao_local'].str.split()
dados.head()
```

```python
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"','',regex=True)
```

```python
dados['comodidades'].str.split(',')
```

```python
dados.head()
```
