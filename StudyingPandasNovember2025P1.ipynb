# Manipulando os dados

## Criando colunas numéricas

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
dados.head()
```

```python
dados['ValorPorMes'] = dados['Valor'] + dados['Condominio']
dados.head()
```

```python
dados['ValorAnual'] = dados['ValorPorMes'] * 12 + dados['IPTU']
dados.head()
```

## Criando colunas categóricas

```python
dados['Descricao'] = dados['Tipo'] + ' no bairro ' + dados['Bairro'] + \
                     ' com ' + dados['Quartos'] + ' Quarto(s) ' \
                     + dados['Vagas'] + 'vaga(s) de garagem. '
```

```python
dados['Descricao'] = dados['Tipo'] + ' no bairro ' + dados['Bairro'] + \   # Concatenando as strings das colunas e convertendo para string o que não era com o (.astype(str)) #
                     ' com ' + dados['Quartos'].astype(str) + ' Quarto(s) e ' \
                     + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
dados.head()
```

```python
dados['possuiSuite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não") # Aplica uma função lambda especificada a cada linha do código #
dados.head()
```

```python
dados.to_csv('dadosCompletosDev.csv', index=False, sep=';')
```
